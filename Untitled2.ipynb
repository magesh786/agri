```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
from PIL import Image
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
train_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

test_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
# Create class subdirectories for the test dataset
!mkdir -p /content/sample_data/image/test/cjd_class
import os
import shutil

# Check if CJD.png exists before moving
if os.path.exists('/content/sample_data/image/test/CJD.png'):
    shutil.move('/content/sample_data/image/test/CJD.png', '/content/sample_data/image/test/cjd_class/')
else:
    print("CJD.png not found in /content/sample_data/image/test, assuming it was already moved.")

# Create class subdirectories for the train dataset
!mkdir -p /content/sample_data/image/train/medical_images

source_dir = "/content/sample_data/image/train"
destination_dir = "/content/sample_data/image/train/medical_images"

# Ensure the source_dir for moving files is the parent of the new class directory
# This loop should only move files that are directly in source_dir, not the new 'medical_images' dir itself
for filename in os.listdir(source_dir):
    full_path = os.path.join(source_dir, filename)
    # Check if it's a file and not the newly created subdirectory
    if os.path.isfile(full_path):
        shutil.move(full_path, destination_dir)


train_dataset = datasets.ImageFolder("/content/sample_data/image/train", transform=train_transform)
test_dataset  = datasets.ImageFolder("/content/sample_data/image/test", transform=test_transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader  = DataLoader(test_dataset, batch_size=32, shuffle=False)

class_names = train_dataset.classes
num_classes = len(class_names)

print("Train Classes:", class_names)
print("Test Classes:", test_dataset.classes)
```

```python
!ls -R /content/sample_data/image
```

```python
model = models.efficientnet_b0(pretrained=True)

# Replace classifier for your dataset
model.classifier[1] = nn.Linear(
    model.classifier[1].in_features,
    num_classes
)

model = model.to(device)
```

```python
for param in model.features.parameters():
    param.requires_grad = False
```

```python
criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(
    model.classifier.parameters(),
    lr=0.001
)
```

```python
epochs = 5

for epoch in range(epochs):
    model.train()
    running_loss = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{epochs}] Loss: {running_loss:.4f}")
```

```python
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        _, predicted = torch.max(outputs, 1)

        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f"Test Accuracy: {accuracy:.2f}%")
```

```python
image = Image.open("/content/sample_data/normal.webp").convert("RGB")

transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

image = transform(image).unsqueeze(0).to(device)

model.eval()
with torch.no_grad():
    output = model(image)
    _, pred = torch.max(output, 1)

print("Predicted class:", class_names[pred.item()])
```

```python
def check_disease(image_path):
    image = Image.open(image_path).convert("RGB")
    image = test_transform(image)
    image = image.unsqueeze(0).to(device)  # Add batch dimension

    with torch.no_grad():
        output = model(image)
        probs = torch.softmax(output, dim=1)
        confidence, predicted_class = torch.max(probs, 1)

    class_names = train_dataset.classes
    disease_status = class_names[predicted_class.item()]

    return disease_status, confidence.item()
```

```python
image_path = "/content/sample_data/normal.webp"

result, confidence = check_disease(image_path)

print("Diagnosis:", result)
print(f"Confidence: {confidence*100:.2f}%")
```
